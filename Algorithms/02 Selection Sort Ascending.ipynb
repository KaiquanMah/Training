```python
arr_sorted = arr_0

for i in range(0, len(arr_sorted)):
    item_i = arr_sorted[i]
    print(f"Loop {i+1}. item_i: {item_i}. Current Array: {arr_sorted}")
    
    
    for j in range(i+1,len(arr_sorted)):
        item_i = arr_sorted[i]
        item_j = arr_sorted[j]
        if item_j < item_i:
            arr_sorted[i] = item_j
            arr_sorted[j] = item_i
    
    print("\n")

    
    
    
arr_sorted
```
