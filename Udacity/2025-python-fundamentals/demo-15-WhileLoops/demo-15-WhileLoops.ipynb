### Coding Practice

Write a program that asks the user to input a positive integer n and then calculates the sum of all natural numbers from 1 to n using a while loop.

Hint: Use a loop to repeatedly add numbers to a total sum variable.

```python
user_num = int(input("Please input a positive integer"))

num = 1
total = 0
while num < user_num+1:
    total+=num
    num+=1
print(f"Sum of numbers from 1 to {user_num}: {total}")
```

### Coding Practice

Write a program that asks the user for a number and then prints the multiplication table for that number up to 10, using a while loop.

Hint: Use a loop to multiply the number by 1, 2, 3, ..., 10.

```python
user_num = int(input("Please input a positive integer, then I will output the multiplication table from 1 to 10"))

num = 1
while num < user_num+1:
    print(f"{num} * {user_num} = {num*user_num}")
    num+=1
```
