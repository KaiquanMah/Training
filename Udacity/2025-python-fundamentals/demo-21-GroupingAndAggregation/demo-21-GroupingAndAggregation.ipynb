```python
# Define the mapping dictionary
priority_mapping = {
    'Critical': 0,
    'High': 1,
    'Medium': 2,
    'Low': 3
}

# Create a new column "Order Ranking" based on the mapping
superstore_df['Order Ranking'] = superstore_df['Order Priority'].map(priority_mapping)

superstore_df['Order Priority'].value_counts()
```

```python
# Define the mapping dictionary
priority_mapping = priority_mapping = {
    'Critical': 0,
    'High': 1,
    'Medium': 2,
    'Low': 3
}

# Apply the mapping to the 'Order Priority' column
superstore_df['Order Priority'] = superstore_df['Order Priority'].replace(priority_mapping)

superstore_df['Order Priority'].value_counts()
```

```python
# Group the DataFrame by 'Segment' and perform aggregation
segment_group = superstore_df_subset.groupby('Segment').agg(
    {'Sales': 'sum', 'Profit': 'sum', 'Discount': 'mean'}
)

# Define the column renaming mapping
column_mapping = {
    'Sales': 'Sum of Sales',
    'Profit': 'Sum of Profit',
    'Discount': 'Average of Discount'
}

# Rename the columns using the mapping
segment_group = segment_group.rename(columns=column_mapping)

segment_group
```
