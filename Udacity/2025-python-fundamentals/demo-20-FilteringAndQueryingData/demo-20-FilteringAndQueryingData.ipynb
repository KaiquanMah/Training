```python
import pandas as pd
```

```python
sales_df = pd.read_csv('datasets/SaaS_Sales.csv')

sales_df = sales_df.drop(
    columns=['Row ID', 'Date Key', 'License'],
    axis = 1
)

sales_df.head().T
```

```python
sales_df.index = sales_df['Order ID']

sales_df = sales_df.drop(
    columns=['Order ID'],
    axis = 1
)

sales_df.head().T
```

```python
sales_df['Order Date'] = pd.to_datetime(sales_df['Order Date'])

sales_df.info()
```

```python
sales_df_filtered = sales_df.filter(like='Customer')

sales_df_filtered.head()
```

```python
sales_df_filtered = sales_df.filter(items=['Product', 'Sales'])

sales_df_filtered.head()
```

```python
sales_df_filtered = sales_df.filter(like='AMER', axis = 0)

sales_df_filtered.head()
```

```python
sales_df_filtered = sales_df.filter(like='2021', axis = 0)

sales_df_filtered.head()
```

```python
india_sales = sales_df[sales_df['Country'] == 'India']

india_sales.head()
```

```python
filtered_sales = sales_df[sales_df['Country'].isin(['India', 'Japan', 'Spain'])]

filtered_sales.sample(5)
```

```python
filtered_sales = sales_df.query('Country in ["India", "Japan", "Spain"]')

filtered_sales.sample(5)
```

```python
filtered_sales = sales_df[sales_df['Segment'].isin(['SMB', 'Strategic'])]

filtered_sales.head()
```

```python
filtered_sales = sales_df.query('Segment in ["SMB", "Strategic"]')

filtered_sales.head()
```

```python
filtered_sales = sales_df[~sales_df['Segment'].isin(['SMB', 'Strategic'])]

filtered_sales.head()
```

```python
filtered_sales = sales_df.query('Segment not in ["SMB", "Strategic"]')

filtered_sales.head()
```

```python
profitable_sales = sales_df[sales_df['Profit'] > 0]

profitable_sales.head()
```

```python
profitable_sales = sales_df.query('Profit > 0')

profitable_sales.head()
```

```python
average_discount = sales_df['Discount'].mean()

discounted_sales = sales_df[sales_df['Discount'] > average_discount]

print('Average discount: ', average_discount)

discounted_sales.head()
```

```python
# Calculate average discount
average_discount = sales_df['Discount'].mean()

# Filter all records where discount was greater than the average discount
filtered_sales = sales_df.query('Discount > @average_discount')

filtered_sales.head()
```

```python
profitable_sales = sales_df[(sales_df['Profit'] > 0) & (sales_df['Discount'] == 0)]

profitable_sales.head()
```

```python
filtered_sales = sales_df.query('Profit > 0 and Discount == 0')

filtered_sales.head()
```

```python
# Calculate average sales
average_sales = sales_df['Sales'].mean()

# Filter all records which are profitable and where sales was greater than the average sales
profitable_sales = sales_df[(sales_df['Profit'] > 0) & (sales_df['Sales'] > average_sales)]

print('Average sales: ', average_sales)

profitable_sales.head()
```

```python
# Calculate average sales
average_sales = sales_df['Sales'].mean()

# Filter all records which are profitable and where sales was greater than the average sales
filtered_sales = sales_df.query('Profit > 0 and Sales > @average_sales')

filtered_sales.head()
```

```python
# Calculate average sales
average_sales = sales_df['Sales'].mean()

# Filter all records which are profitable or where sales was greater than the average sales
profitable_sales = sales_df[(sales_df['Profit'] > 0) | (sales_df['Sales'] > average_sales)]

print('Average sales: ', average_sales)

profitable_sales.head()
```

```python
# Calculate average sales
average_sales = sales_df['Sales'].mean()

# Filter all records which are profitable and where sales was greater than the average sales
filtered_sales = sales_df.query('Profit > 0 or Sales > @average_sales')

filtered_sales.head()
```

```python
# Filter all records where the region is 'AMER' or the the customer is 'Royal Dutch Shell'
filtered_sales = sales_df[(sales_df['Region'] == 'AMER') | (sales_df['Customer'] == 'Royal Dutch Shell')]

filtered_sales.head()
```

```python
filtered_sales = sales_df.query('Region == "AMER" or Customer == "Royal Dutch Shell"')

filtered_sales.head()
```
