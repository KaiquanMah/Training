Creating dataframes

```python
sales_data = [ 
    {'Id' : 1, 'Product': 'Phone', 'Price': 500},
    {'Id' : 2, 'Product': 'Laptop', 'Price': 1200},
    {'Id' : 3, 'Product': 'Tablet', 'Price': 800},
    {'Id' : 4, 'Product': 'Headphones', 'Price': 100}
]
```

```python
sales_df = pd.DataFrame(sales_data)

sales_df
```

```python
products = ['Phone', 'Laptop', 'Tablet', 'Headphones']

quantities = [12, 5, 8, 15]
```

```python
product_data = list(zip(products, quantities))

product_data
```

```python
product_df = pd.DataFrame(product_data, columns=['Product', 'Quantity'])

product_df
```
